```python
import networkx as nx
from nxpd import draw
from nxpd import nxpdParams
nxpdParams['show'] = 'ipynb'
```

```python
dot = nx.DiGraph()
dot.add_node('A', {'label':'(Random) Starting points (.rst + .prmtop)'})
dot.add_node('B', {'label':'PBS script'})
dot.add_node('C', {'label':'MD simulation', 
                   'color':'blue'})
dot.add_node('D', {'label': 'MSM',
                   'color':'red'})
dot.add_node('Check', {'label':'Number of max Epochs reached?'})
dot.add_node('E', {'label':'New Starting points (.rst)',
                   'color':'red'})
dot.add_node('F', {'label':'End','shape':'doublecircle'})
dot.add_node('H', {'label':'Store Data (.nc)'})
dot.add_edges_from([('A', 'B', {'label':'Epoch n=1'}),
                   ('B', 'C', {'label':'Inputs (.pbs, .rst, .prmtop, .in)'}),
                   ('D', 'E', {'label':'Undersampled States'}),
                   ('C', 'Check'),
                   ('Check', 'D', {'label':'No'}),
                   ('Check', 'F', {'label':'Yes'}),
                   ('E', 'B', {'label':'Next Epoch n+=1'}),
                   ('C', 'H',{'label':'Store Data (.nc)'}),
                   ('H', 'F', {'style':'dashed',
                               'label':'Final Data Set (.nc)'})])
```

```python
draw(dot)
```
